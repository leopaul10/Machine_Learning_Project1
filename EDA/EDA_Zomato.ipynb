```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how="left")
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
values=final_df.Country.value_counts().values
```

```python
plt.pie(values,labels = names)
```

```python
plt.pie(values,labels = names)
```

```python
plt.pie(values[:3],labels = names[:3],radius=1.5,autopct= '%1.2f%%')
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings.head()
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating)
```

```python
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
```
